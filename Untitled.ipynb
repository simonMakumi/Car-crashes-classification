## Data Understanding

### Sources
The datasets used in this project are:
1. **People Dataset**: Information about people involved in crashes.
2. **Vehicle Dataset**: Information about vehicles involved in crashes.

Both datasets are available for download from the Chicago Data Portal.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the People dataset in chunks to avoid errors when loading due to data size.
chunk_size = 50000
chunks = pd.read_csv('Traffic_Crashes_-_People_20240531.csv', chunksize=chunk_size)

# Initialize an empty list to store the chunks
people_chunks = []

# Iterate over the chunks and append them to the list
for chunk in chunks:
    people_chunks.append(chunk)

# Concatenate the chunks to create the full People dataset
people_data = pd.concat(people_chunks, ignore_index=True)

# Display the first few rows of the People dataset
people_data.head()
```

```python
# Loading the vehicle dataset and display.
vehicle_data = pd.read_csv('Traffic_Crashes_-_Vehicles_20240531.csv')
vehicle_data.head()
```

## Data Preparation

In this section, we merge the People and Vehicle datasets based on the `CRASH_RECORD_ID` column. This will allow us to combine information about people involved in crashes with information about the vehicles involved.


```python
# Merge the People and Vehicle datasets based on CRASH_RECORD_ID
df = pd.merge(people_data, vehicle_data, on='CRASH_RECORD_ID', how='inner')

# Display the first few rows of the merged dataset
df.head()
```

```python
df.columns
```

```python
relevant_columns = ['PERSON_TYPE', 'SEX', 'AGE', 'SAFETY_EQUIPMENT','AIRBAG_DEPLOYED', 'EJECTION', 'INJURY_CLASSIFICATION',
                    'DRIVER_ACTION', 'DRIVER_VISION', 'PHYSICAL_CONDITION', 'BAC_RESULT VALUE', 'CELL_PHONE_USE',
                    'NUM_PASSENGERS', 'MAKE', 'MODEL', 'VEHICLE_YEAR', 'VEHICLE_DEFECT','VEHICLE_USE', 'CRASH_DATE_x']
df = df[relevant_columns]
df.head()
```

```python
df.info()
```

# Data Cleaning: Handling Missing Values

In this step, we address missing values in the dataset. Missing values can hinder analysis and modeling efforts, so it's essential to handle them appropriately. We employ several strategies to deal with missing values, including filling with placeholders, dropping rows or columns, and imputing with statistical measures.

```python
df.isna().sum()
```

## 1. Fill Missing Values

### Fill with 'UNKNOWN'

Certain columns contain categorical data where missing values can be reasonably replaced with 'UNKNOWN'. These columns include:
- `DRIVER_ACTION`
- `DRIVER_VISION`
- `VEHICLE_DEFECT`
- `VEHICLE_USE`
- `PHYSICAL_CONDITION`

### Drop 'rows'

The following rows with missing values will be dropped:
- `SAFETY_EQUIPMENT`
- `AIRBAG_DEPLOYED`
- `EJECTION`
- `INJURY_CLASSIFICATION`
- `MAKE`
- `MODEL`

### Drop columns with too many missing values
The following columns will be affected:
- `BAC_RESULT VALUE`
- `CELL_PHONE_USE`
- `NUM_PASSENGERS`

### Fill with median and SEX with X
These columns will be filled with the median:
- `AGE`
- `VEHICLE_YEAR`
- `SEX`

```python
# Fill with 'UNKNOWN'
unknown_columns = ['DRIVER_ACTION', 'DRIVER_VISION', 'VEHICLE_DEFECT', 'VEHICLE_USE', 'PHYSICAL_CONDITION']
df[unknown_columns] = df[unknown_columns].fillna('UNKNOWN')

# Drop rows with missing values in these columns
drop_rows_columns = ['SAFETY_EQUIPMENT', 'AIRBAG_DEPLOYED', 'EJECTION', 'INJURY_CLASSIFICATION', 'MAKE', 'MODEL']
df.dropna(subset=drop_rows_columns, inplace=True)

# Drop columns with too many missing values
drop_columns = ['BAC_RESULT VALUE', 'CELL_PHONE_USE', 'NUM_PASSENGERS']
df.drop(columns=drop_columns, inplace=True)

# Fill with median
df['AGE'].fillna(df['AGE'].median(), inplace=True)
df['SEX'].fillna('X', inplace=True)
df['VEHICLE_YEAR'].fillna(df['VEHICLE_YEAR'].median(), inplace=True)

# Check the dataset after handling missing values
print(df.isnull().sum())

# Display the first few rows of the cleaned dataset
df.head()
```
